```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def openFile(path):
    data = None
    with open(path, 'r') as f:
        data = f.read()
        f.close()

    data = data.replace('\n', '')

    ree = np.zeros(len(data))

    for i, d in enumerate(data):
        ree[i] = d
    return ree
```

```python
data_a = {
    'raw': openFile("data_RandomDigits2025_A.txt"),
    'name': 'A'
}
data_b = {
    'raw': openFile("data_RandomDigits2025_B.txt"),
    'name': 'B'
}
data_c = {
    'raw': openFile("data_RandomDigits2025_C.txt"),
    'name': 'C'
}
data_d = {
    'raw': openFile("data_RandomDigits2025_D.txt"),
    'name': 'D'
}
data_e = {
    'raw': openFile("data_RandomDigits2025_E.txt"),
    'name': 'E'
}

data = [data_a, data_b, data_c, data_d, data_e]
```

```python
def countDigits(rawData):
    counts = {}
    for d in rawData:
        if str(d) not in counts:
            counts[str(d)] = 0
        counts[str(d)] += 1
    return counts
```

```python
for d in data:
    d['counts'] = countDigits(d['raw'])
    plt.hist(d['raw'], bins = 10, range = (0, 9), histtype='step', label = f'Data {d['name']}');
    plt.show()
plt.legend();
```

```python
for d in data:
    
    plt.hist(d['raw'][1:] - d['raw'][:-1], bins = 10)
    plt.show()
```

```python
for d in data:
    
    plt.hist(d['raw'][:-1] - d['raw'][1:], bins = 10)
    plt.show()
```

```python
digits = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
for d in data:
    res = []
    for digit in digits:

        r = 0
        for i in range(len(d['raw']) - 1):
            if d['raw'][i] == digit and d['raw'][i+1] == digit:
                r += 1
        
        res.append(r)
    print(res)
    
    plt.bar(digits, res)
    plt.xlabel("Digit")
    plt.ylabel("Count of consecutive duplicates")
    plt.show()
```

```python
digits = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
for d in data:
    res = []
    for digit in digits:

        r = 0
        for i in range(len(d['raw']) - 2):
            if d['raw'][i] == digit and d['raw'][i+1] == digit and d['raw'][i+2] == digit:
                r += 1
        
        res.append(r)

    print(res, len(d['raw']))
    
    plt.bar(digits, res)
    plt.xlabel("Digit")
    plt.ylabel("Count of consecutive duplicates")
    plt.show()
```
